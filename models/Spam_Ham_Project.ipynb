```python
#Importing libraries
import pandas as pd
file_path = '/content/spam.csv'
messages = pd.read_csv(file_path, delimiter=',', encoding='latin1', header=None)
messages = messages.rename(columns = {0:'label', 1:'message'})
messages.head()
```

```python
#Importing the regular expression and nltk libraries
import re
import nltk
nltk.download('stopwords')
```

```python
#Calling the porterstemmer
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
#Removing the stopwords
corpus = []
for i in range(0,len(messages)):
  review = re.sub('[^a-zA-Z]',' ',messages['message'][i])
  review = review.lower()
  review = review.split()
  review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
  review = ' '.join(review)
  corpus.append(review)
```

```python
corpus
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=100,ngram_range=(1,2))
```

```python
X = cv.fit_transform(corpus).toarray()
```

```python
import numpy as np
np.set_printoptions(edgeitems=30, linewidth=100000, formatter=dict(float=lambda x: "%.3g" % x))
X
```

```python
cv.vocabulary_
```

```python
y = pd.get_dummies(messages['label'])
```

```python
y = y.iloc[:,0].values
```

```python
#Splitting the dataset into train and test datasets
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3)
```

```python
from sklearn.naive_bayes import MultinomialNB
spam_detect_model = MultinomialNB().fit(X_train,y_train)
```

```python
y_pred = spam_detect_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
import nltk
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
corpus = []
for i in range(0,len(messages)):
  review = re.sub('[^a-zA-Z]',' ',messages['message'][i])
  review = review.lower()
  review = review.split()
  review = [lemmatizer.lemmatize(word) for word in review ]
  review = ' '.join(review)
  corpus.append(review)
```

```python
from nltk import sent_tokenize
!pip install --upgrade gensim # upgrade gensim to the latest version
from gensim.utils import simple_preprocess # import simple_preprocess from gensim.utils
```

```python
words = []
for sent in corpus:
  sent_token = sent_tokenize(sent)
  for sent in sent_token:
    words.append(simple_preprocess(sent))
```

```python
def avg_word2vec(doc):
  #remove out of vocabulary words
  #sent = [word for word in doc if word in in model.wv.index_to_key]
  #print(sent)

  return np.mean([model.wv[word] for word in doc if word in model.wv.index_to_key])
```
